#  Matriz DFT (Transformada Discreta de Fourier)

As definições da $\mathbf{DFT}$ (Transformada Discreta de Fourier) e da $\mathbf{IDFT}$ (Transformada Inversa Discreta de Fourier) são apresentadas a seguir.

### Transformada Discreta de Fourier (DFT)

A DFT é definida pela soma:
$$
X(k) = \sum_{n=0}^{N-1} x(n) W_N^{kn}, \quad \text{para } 0 \le k \le N-1
$$

Onde:
* $X(k)$ é o **coeficiente da DFT** (domínio da frequência).
* $x(n)$ é a **amostra do sinal** (domínio do tempo).
* $N$ é o **número de amostras**.
* $W_N = e^{-j\frac{2\pi}{N}}$ é o **fator $W_N$ (twiddle factor)**.

### Transformada Inversa Discreta de Fourier (IDFT)

A IDFT reconstrói o sinal original:
$$
x(n) = \frac{1}{N} \sum_{k=0}^{N-1} X(k) W_N^{-kn}, \quad \text{para } 0 \le n \le N-1
$$

Onde:
* $x(n)$ é a **amostra reconstruída**.
* $X(k)$ é o **coeficiente da DFT**.
* $W_N^{-1} = e^{+j\frac{2\pi}{N}}$ (o conjugado complexo de $W_N$).

##  Notação Matricial



As transformações DFT e IDFT de um único coeficiente podem ser expressas como produtos vetoriais.

### DFT como Produto Vetorial (Linha $k$)

O $k$-ésimo coeficiente da DFT, $X(k)$, é o produto interno entre a $k$-ésima linha da matriz DFT e o vetor do sinal $x$:
$$
X(k) =
\begin{bmatrix}
W_N^0 & W_N^k & W_N^{2k} & \dots & W_N^{(N-1)k}
\end{bmatrix}
\begin{bmatrix}
x(0) \\ x(1) \\ x(2) \\ \vdots \\ x(N-1)
\end{bmatrix}
$$

### IDFT como Produto Vetorial (Linha $n$)

A $n$-ésima amostra reconstruída, $x(n)$, é o produto interno (escalonado por $1/N$) entre a $n$-ésima linha da matriz IDFT e o vetor dos coeficientes $X$:
$$
x(n) = \frac{1}{N}
\begin{bmatrix}
W_N^0 & W_N^{-n} & W_N^{-2n} & \dots & W_N^{-(N-1)n}
\end{bmatrix}
\begin{bmatrix}
X(0) \\ X(1) \\ X(2) \\ \vdots \\ X(N-1)
\end{bmatrix}
$$

## Forma Matricial Completa ($N \times N$)

A $\mathbf{DFT}$ de todo o vetor $\mathbf{x}$ resulta no vetor $\mathbf{X}$ através da multiplicação pela matriz $\mathbf{W}_N$:
$$
\begin{bmatrix}
X(0) \\ X(1) \\ X(2) \\ \vdots \\ X(N-1)
\end{bmatrix}
=
\begin{bmatrix}
W_N^0 & W_N^0 & \dots & W_N^0 \\
W_N^0 & W_N^1 & \dots & W_N^{N-1} \\
W_N^0 & W_N^2 & \dots & W_N^{2(N-1)} \\
\vdots & \vdots & \ddots & \vdots \\
W_N^0 & W_N^{N-1} & \dots & W_N^{(N-1)^2}
\end{bmatrix}
\begin{bmatrix}
x(0) \\ x(1) \\ x(2) \\ \vdots \\ x(N-1)
\end{bmatrix}
$$


### IDFT

A $\mathbf{IDFT}$ de todo o vetor $\mathbf{X}$ resulta no vetor $\mathbf{x}$ através da multiplicação pela matriz conjugada e escalonada:
$$
\begin{bmatrix}
x(0) \\ x(1) \\ x(2) \\ \vdots \\ x(N-1)
\end{bmatrix}
=
\frac{1}{N}
\begin{bmatrix}
W_N^0 & W_N^0 & \dots & W_N^0 \\
W_N^0 & W_N^{-1} & \dots & W_N^{-(N-1)} \\
W_N^0 & W_N^{-2} & \dots & W_N^{-2(N-1)} \\
\vdots & \vdots & \ddots & \vdots \\
W_N^0 & W_N^{-(N-1)} & \dots & W_N^{-(N-1)^2}
\end{bmatrix}
\begin{bmatrix}
X(0) \\ X(1) \\ X(2) \\ \vdots \\ X(N-1)
\end{bmatrix}
$$

##  Forma Compacta e Propriedades



### Definição dos Vetores e Matriz

* **Vetores de Entrada e Saída:**
    $$
    \mathbf{x} =
    \begin{bmatrix}
    x(0) \\ x(1) \\ \vdots \\ x(N-1)
    \end{bmatrix},
    \quad
    \mathbf{X} =
    \begin{bmatrix}
    X(0) \\ X(1) \\ \vdots \\ X(N-1)
    \end{bmatrix}
    $$

* **Matriz da DFT ($\mathbf{W}_N$):** O elemento na linha $i$ e coluna $j$ (começando em 0) é:
    $$
    \{\mathbf{W}_N\}_{i,j} = W_N^{ij}, \quad \text{para } 0 \le i,j \le N-1
    $$

### Expressões Compactas

* **DFT:**
    $$
    \mathbf{X} = \mathbf{W}_N \mathbf{x}
    $$

* **IDFT:**
    $$
    \mathbf{x} = \frac{1}{N} \mathbf{W}_N^{*} \mathbf{X}
    $$

### Propriedades da Matriz $\mathbf{W}_N$

* A matriz $\mathbf{W}_N$ é **simétrica**: $\mathbf{W}_N^T = \mathbf{W}_N$.
* A **inversa** da matriz $\mathbf{W}_N$ é dada por: $\mathbf{W}_N^{-1} = \frac{1}{N} \mathbf{W}_N^*$.
* **Complexidade Computacional (implementação direta):**
    * Número de multiplicações complexas não triviais: $N^2 - 2N + 1$.
    * Número total de somas complexas: $N(N-1)$. (Note que a FFT reduz significativamente essa complexidade).

### Implementando a DFT de forma Matricial

Construção da Matriz de Fourier ($W$)O cerne da mudança está aqui. A DFT pode ser vista como o produto de uma matriz especial, a Matriz de Fourier ($W$), pelo vetor do sinal ($x$).Matriz de Fourier ($W$): É uma matriz $N \times N$ onde o elemento na linha $k$ e coluna $n$ é dado por:$$W_{k,n} = e^{-i 2 \pi k n / N}$$
- Implementação:Criamos os vetores n e k. É crucial que k seja um vetor coluna (matriz $N \times 1$) e n seja um vetor linha (matriz $1 \times N$). Isso é feito com k = n.reshape((N, 1)). A multiplicação $k \cdot n$ resulta automaticamente em uma matriz $N \times N$, onde cada elemento $(k, n)$ é o produto $k \cdot n$. 
- O NumPy aplica a função exponencial (np.exp) em todos os elementos da matriz resultante, criando a matriz $W$ em uma única operação.

#### Multiplicação Matricial
- A DFT, $X$, é o resultado da multiplicação da Matriz de Fourier $W$ pelo vetor de sinal $\text{sig}$:$$X = W \cdot \text{sig}$$Implementação: Usamos o operador @ (ou np.dot), que realiza a multiplicação matricial: X = W @ sig.
Esta única linha de código substitui os seus dois laços aninhados, resultando em uma execução muito mais rápida para grandes valores de $N$ porque o NumPy utiliza implementações de baixo nível (muitas vezes em C ou Fortran) otimizadas para álgebra linear.

### Observações :

-  A função da DFT realiza o zero-padding para N > L.
- A funçao da DFT realiza o truncamento para L > N.
- Realiza multiplicação Linha x Coluna.

```python

def DFT_func_matricial(signal, N):
    """
    Calcula a Transformada Discreta de Fourier (DFT) de forma matricial.

    Args:
        signal (array-like): O sinal de entrada.
        N (int): O número de pontos da DFT.

    Returns:
        numpy.ndarray: O vetor da DFT normalizada.
    """
    L = len(signal)
    
    # 1. Preparação/Ajuste do Sinal (Zero Padding ou Truncamento)
    sig = np.zeros(N, dtype=signal.dtype)
    if L < N:
        sig[:L] = signal  # Zero Padding
    else:
        sig = signal[0:N] # Truncamento
        
    # Certifica-se de que o tipo de dado é complexo para o cálculo subsequente
    sig = sig.astype(np.complex128)

    # 2. Construção da Matriz de Fourier (W)
    # n e k variam de 0 a N-1.
    n = np.arange(N)
    k = n.reshape((N, 1)) # Transpõe 'n' para que 'k' seja uma coluna (N x 1)

    # O expoente é: -2j * pi * k * n / N
    # k e n são vetores, k*n resultará em uma matriz (N x N)
    exponent = (-2j * np.pi * k * n) / N
    
    # A Matriz de Fourier (W) é calculada como e^(expoente)
    W = np.exp(exponent)
    # 

    # 3. Multiplicação Matricial
    # X[k] = sum_{n=0}^{N-1} sig[n] * W[k, n]
    # Isso é exatamente o produto da Matriz W (N x N) pelo vetor sig (N x 1)
    X = W @ sig 
    # Alternativamente: X = np.dot(W, sig)

    # 4. Normalização
    return X / np.sqrt(N)
```

### Uso da GPU via CuPy para via de comparação com a CPU, em velocidade de processamento.

```python
def DFT_func_cupy(signal_np, N):
    """
    Calcula a Transformada Discreta de Fourier (DFT) de forma matricial usando CuPy.

    Args:
        signal_np (numpy.ndarray): O sinal de entrada (deve ser um array NumPy na CPU).
        N (int): O número de pontos da DFT.

    Returns:
        numpy.ndarray: O vetor da DFT normalizada (copiado de volta para a CPU).
    """
    # 1. Transferência de Dados: CPU -> GPU
    # Move o sinal de entrada da memória da CPU para a memória da GPU
    signal_cp = cp.asarray(signal_np)

    # 2. Preparação/Ajuste do Sinal (Executado na GPU)
    L = len(signal_cp)
    sig_cp = cp.zeros(N, dtype=signal_cp.dtype)
    
    if L < N:
        sig_cp[:L] = signal_cp  # Zero Padding
    else:
        # Apenas garantir o truncamento e tipo complexo na GPU
        sig_cp = signal_cp[0:N].astype(cp.complex128)
    
    # Se já fizemos o ajuste de tamanho, garantimos o tipo complexo
    if sig_cp.dtype != cp.complex128:
         sig_cp = sig_cp.astype(cp.complex128)

    # 3. Construção da Matriz de Fourier (W) na GPU
    # Usamos cp.arange e cp.pi (versões CuPy)
    n = cp.arange(N)
    k = n.reshape((N, 1)) 

    # Cálculo do expoente na GPU
    exponent = (-2j * cp.pi * k * n) / N
    
    # A Matriz de Fourier (W) é criada na memória da GPU
    W = cp.exp(exponent)

    # 4. Multiplicação Matricial (Onde o ganho de velocidade é máximo)
    # A multiplicação é feita de forma paralela e altamente eficiente na GPU
    X_cp = W @ sig_cp

    # 5. Normalização na GPU
    X_cp /= cp.sqrt(N)
    
    # 6. Transferência de Dados: GPU -> CPU
    # Move o resultado de volta para a memória da CPU para uso posterior
    return cp.asnumpy(X_cp)
```

## DFT de forma Iterativa e Usando decorador Numba para Aceleração de Código

```python
@njit(parallel=True)
def DFT_func(signal, N):
    L = len(signal)  # tamanho do sinal

    sig = np.zeros(N, dtype=signal.dtype) # inicializa o vetor do sinal com zeros de tamanho N do tipo do sinal
    if L < N:
        sig[:L] = signal  # zero padding
    else:
        sig = signal[0:N]  # truncar

    X = np.zeros(N, dtype=np.complex128)  # inicializa o vetor da DFT com zeros
    for k in prange(N):
        for n in range(N):
            X[k] += sig[n] * (np.exp((-2j * np.pi * k * n) / N))  # soma os valores da DFT

    return X / np.sqrt(N)  # normaliza a DFT
```

```python
L1= 30
N1= 100

n1= np.linspace(-10,10,L1)

x1= np.cos(2*np.pi*n1 + np.pi/4) #cosine signal


X1_DFT= DFT_func(x1,N=N1) #DFT do sinal x1, L1=50 < N1=100, com zero padding

X2_DFT= DFT_func_matricial(x1,N=N1) #DFT do sinal x2, L2=100 > N2=80, truncado

X3_DFT= DFT_func(x1,N=1000) #DFT do sinal x2, L2=100 < N=1000

X4_DFT= DFT_func_matricial(x1,N=1000) #DFT do sinal x2, L2=100 > N=1000

X5_DFT= DFT_func_cupy(x1,N=N1) #DFT do sinal x2, L2=100 > N=1000 usando CuPy

X6_DFT= DFT_func_cupy(x1,N=1000) #DFT do sinal x2, L2=100 > N=1000 usando CuPy

fig , axs= plt.subplots(3,3, figsize=(20,15))

axs[0,0].stem(n1,x1)
axs[0,0].set_title('Sinal x1[n]')
axs[1,0].stem(n1,x1)
axs[1,0].set_title('Sinal x1[n]')
axs[0,1].stem(np.abs(X1_DFT))
axs[0,1].set_title('Magnitude DFT x1[n] DFT Iterativa ')
axs[1,1].stem(np.abs(X2_DFT))
axs[1,1].set_title('Magnitude DFT x1[n] DFT Matricial ')
axs[0,2].stem(np.abs(X3_DFT))
axs[0,2].set_title('Magnitude DFT x1[n]- N=1000')
axs[1,2].stem(np.abs(X4_DFT))
axs[1,2].set_title('Magnitude DFT x1[n]- N=1000')
axs[2,0].stem(n1,x1)
axs[2,0].set_title('Sinal x1[n] - CuPy')
axs[2,1].stem(np.abs(X5_DFT))
axs[2,1].set_title('Magnitude DFT x1[n] com zero padding - CuPy')
axs[2,2].stem(np.abs(X6_DFT))
axs[2,2].set_title('Magnitude DFT x1[n] - N=1000')
plt.show()
```

## Teste de desempenho: DFT Iterativa x DFT Matricial x DFT Matricial (CuPy) x FFT(NumPy)

```python
N_values = [2**i for i in range(5, 16)] # N = 32, 64, ..., 2048 (para teste rápido)
# N_values = [2**i for i in range(5, 14)] # Se tiver tempo, aumente N para ~8192

# Inicializa listas para armazenar os tempos
times_dft_matrix = []
times_dft_cp = []
times_fft_np = []
times_dft_for = []

print("Iniciando Benchmark...")

for N in N_values:
    # Cria um sinal de exemplo (real)
    signal = np.random.rand(N) + np.cos(2 * np.pi * 0.1 * np.arange(N))  # Sinal com componente senoidal
    
    #DFT Iteraiva

    start_time = time.time()
    DFT_func(signal, N)
    end_time = time.time()
    times_dft_for.append(end_time - start_time)

    # --- DFT Matricial (NumPy) ---
    start_time = time.time()
    DFT_func_matricial(signal, N)
    end_time = time.time()
    times_dft_matrix.append(end_time - start_time)
    
    # --- DFT Matricial (CuPy) ---
    # CuPy precisa de uma "pré-execução" (warm-up) para inicializar a GPU
    # Para medições mais precisas, faríamos isso e rodaríamos várias vezes.
    try:
        # Medição:
        start_time = time.time()
        DFT_func_cupy(signal, N)
        # Sincroniza a GPU para garantir que a operação terminou antes de medir o tempo
        cp.cuda.get_current_stream().synchronize() 
        end_time = time.time()
        times_dft_cp.append(end_time - start_time)
    except Exception as e:
        # Se CuPy falhar (sem GPU), registra um tempo alto ou 0
        print(f"CuPy falhou para N={N}: {e}")
        times_dft_cp.append(np.nan) 
        
    # --- FFT (NumPy) ---
    start_time = time.time()
    fft=np.fft.fft(signal, N)
    end_time = time.time()
    times_fft_np.append(end_time - start_time)
    
    print(f"Concluído para N={N}")

print("Benchmark concluído!")

# --- 3. Plotagem dos Resultados ---

plt.figure(figsize=(10, 6))
plt.plot(N_values, times_dft_matrix, 'o-', label='DFT Matricial (Matricial, $O(N^2)$)', color='blue')
plt.plot(N_values, times_dft_cp, 's-', label='DFT Matricial (CuPy/GPU, $O(N^2)$)', color='red')
plt.plot(N_values, times_fft_np, '^-', label='FFT (NumPy, $O(N \log N)$)', color='green')
plt.plot(N_values, times_dft_for, 'x-', label='DFT Iterativa (Numba, $O(N^2)$)', color='orange')

plt.xlabel('Tamanho do Sinal (N)')
plt.ylabel('Tempo de Execução (segundos) - Escala Logarítmica')
plt.title('Comparação de Desempenho de Transformadas')
plt.yscale('log') # Usar escala logarítmica no eixo Y ajuda a visualizar grandes diferenças

# Corrige erro: não passe uma lista para xscale. Usar 'log' e ajustar ticks para os valores de N.
plt.xscale('log') 
plt.xticks(N_values, N_values)  # define ticks nos valores de N (potências de 2)

plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()

#
```
